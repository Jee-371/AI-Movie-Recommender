```python
def map_genre_category(genres):
    if pd.isna(genres):
        return "Unknown"
    
    genres = [g.strip() for g in genres.split(",")]

    # Rule 1: Documentary override
    if "Documentary" in genres:
        return "Documentary"

    # Priority order
    priority = [
        "Action",
        "Comedy",
        "Drama",
        "Crime",
        "Romance",
        "Horror",
        "Thriller",
        "Science Fiction",
        "Adventure",
        "Mystery",
        "Fantasy",
        "Family",
        "Animation",
        "War",
        "Western",
    ]

    main = None
    sub = None

    for p in priority:
        if p in genres:
            if not main:
                main = p
            elif not sub:
                sub = p
                break

    if main and sub:
        return f"{main} {sub}"
    elif main:
        return main
    else:
        return genres[0]
```

```python
movies["genre_category"] = movies["genres"].apply(map_genre_category)
```

```python
from transformers import pipeline
```

```python
pipe = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
```

```python
sug_categories = ['Action Drama', 'Drama Romance']
```

```python
seq  = movies.loc[movies['genre_category'] == 'Action Drama', 'overview'].reset_index(drop=True)[0]
```

```python
pipe(seq, sug_categories)
```

```python
import numpy as np
```

```python
max_index = np.argmax(pipe(seq, sug_categories)['scores'])
max_label = pipe(seq, sug_categories)['labels'][max_index]
max_label
```

```python
def generate_predictions(seq, cat):
    pred = pipe(seq, cat)
    
    max_index = np.argmax(pred['scores'])
    max_label = pred['labels'][max_index]
    
    return max_label
```

```python
from tqdm import tqdm
```

```python
actual_cats = []
pred_cats = []
```

```python
for i in tqdm(range(0, 300)):
    seq = movies.loc[movies['genre_category'] == 'Action Drama', 'overview'].reset_index(drop=True)[i]

    pred = generate_predictions(seq, sug_categories)

    pred_cats.append(pred)
    actual_cats.append('Action Drama')
```

```python
for i in tqdm(range(0, 300)):
    seq = movies.loc[movies['genre_category'] == 'Drama Romance', 'overview'].reset_index(drop=True)[i]

    pred = generate_predictions(seq, sug_categories)

    pred_cats.append(pred)
    actual_cats.append('Drama Romance')
```

```python
print(len(actual_cats))
print(len(pred_cats))
```

```python
pred_df = pd.DataFrame({
    'actual_genre': actual_cats,
    'predicted_genre': pred_cats
})
```

```python
pred_df['correct_predictions'] = np.where(pred_df['actual_genre'] == pred_df['predicted_genre'], 1, 0)
pred_df['correct_predictions'].sum() / len(pred_df)
```

```python
movies_sample = movies[['overview','genre_category']].dropna().sample(100, random_state=42).reset_index(drop=True)
```

```python
def gen_prediction(seq, genres):

    pred = pipe(
        seq,
        genres,
        hypothesis_template="This movie is a {} film."
    )

    return pred['labels'][0]
```

```python
actual_gens = []
pred_gens = []

for i in tqdm(range(len(movies_sample))):

    seq = movies_sample['overview'][i]
    actual = movies_sample['genre_category'][i]

    pred = gen_prediction(seq, genre_categories)

    actual_gens.append(actual)
    pred_gens.append(pred)
```

```python
predict_df = pd.DataFrame({
    'actual_genre': actual_gens,
    'predicted_genre': pred_gens
})
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
accuracy = accuracy_score(predict_df['actual_genre'], predict_df['predicted_genre'])

print("Accuracy:", accuracy)
```

```python
print(classification_report(
    predict_df['actual_genre'],
    predict_df['predicted_genre']
))
```
