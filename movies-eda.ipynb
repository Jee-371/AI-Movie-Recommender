```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
movies = pd.read_csv("/kaggle/input/datasets/asaniczka/tmdb-movies-dataset-2023-930k-movies/TMDB_movie_dataset_v11.csv")
```

```python
movies
```

```python
ax = plt.axes()
sns.heatmap(movies.isna().transpose(), cbar=False, ax=ax)

plt.xlabel("Columns")
plt.ylabel("Missing Values")
plt.show()
```

```python
movies = movies.drop(columns=[
    "backdrop_path",
    "poster_path",
    "homepage",
    "imdb_id",
    "production_companies",
    "production_countries",
    "spoken_languages",
    "adult",
    "status",
    "original_title",
    "tagline",
    "revenue",
    "budget"
], errors="ignore")
```

```python
movies
```

```python
movies.isnull().sum()
```

```python
movies = movies[movies["overview"].notna()]
movies = movies[movies["title"].notna()]
movies = movies[movies["genres"].notna()]
```

```python
movies = movies.drop(columns=[
    "keywords",
    "release_date"
], errors="ignore")
```

```python
movies
```

```python
columns_req = ['vote_average','vote_count','popularity']
core_mat = movies[columns_req].corr(method='spearman')
```

```python
core_mat
```

```python
sns.set_theme(style="white")
plt.figure(figsize=(8, 6))

heatmap = sns.heatmap(
    core_mat,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    cbar_kws={"label": "Spearman correlation"}
)

heatmap.set_title("Correlation heatmap")
plt.show()
```

```python
movies['words_in_overview'] = movies['overview'].str.split().str.len()
```

```python
movies
```

```python
movies.sort_values("words_in_overview", ascending=True)[
    ["title", "words_in_overview"]
].head(200)
```

```python
movies = movies[movies["words_in_overview"] >= 24]
```

```python
movies
```

```python
movies = movies[movies["runtime"] >= 10]
```

```python
movies_zero_rating = movies[movies["vote_average"] == 0]

movies_zero_rating[["title", "vote_average", "vote_count"]].head(20)
```

```python
movies = movies[~((movies["vote_average"] < 3) | (movies["vote_count"] < 8))]
```

```python
movies
```

```python
movies = movies.copy()
```

```python
movies["combined_text"] = movies[
    ["title", "overview", "genres", "original_language"]
].fillna("").agg(", ".join, axis=1)
```

```python
movies
```

```python
movies = movies.drop(columns=["words_in_overview"])
```

```python
movies.to_csv("final_movies_dataset.csv", index=False)
```
